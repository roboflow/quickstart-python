# Roboflow Model Types

Roboflow allows you to apply computer vision to your problem with the following tasks:

* Classification
* Object Detection
* Instance Segmentation
* Semantic Segmentation

In this notebook, we are going to load a model to solve problems in each of these task types. Let's get started!

```python
#log into Roboflow
import roboflow
roboflow.login()
```

## Classification Inference

Classification labels an image as belonging to one of a list of possible categories.

With [this open-source model](https://universe.roboflow.com/mohamed-traore-2ekkp/chest-x-rays-qjmia/model/3), we can identify whether or not a patient x-ray displays signs of pneumonia.



```python
img_url = "https://media.roboflow.com/quickstart/chest_xray.jpeg?updatedAt=1678743714452"
universe_model_url = "https://universe.roboflow.com/mohamed-traore-2ekkp/chest-x-rays-qjmia/model/3"

model = roboflow.load_model(model_url=universe_model_url)
pred = model.predict(img_url, hosted=True)
pred.plot()
#we can also get - pred.json()
```

## Object Detection Inference

Object detection models identify the location of objects of interest in an image. After inference, "bounding boxes" are drawn around predictions to visualize the results of the model. With [this model](https://universe.roboflow.com/brad-dwyer/aerial-solar-panels/model/6), we can find the location and count of solar panels from drone imagery.

```python
img_url = "https://media.roboflow.com/quickstart/aerial_drone.jpeg?updatedAt=1678743716455"
universe_model_url = "https://universe.roboflow.com/brad-dwyer/aerial-solar-panels/model/6"

model = roboflow.load_model(model_url=universe_model_url)
pred = model.predict(img_url, hosted=True)
pred.plot()
```

```python
print(len(pred.json()["predictions"]), "solar panels were detected in the image")
```

## Instance Segmentation Inference

Instance segmentation models identify an exact boundary around an object of interest in an image. While more precise than object detection, instance segmentation models are more difficult to create; they require more detailed annotations.

Let's use a model to [find the outline of a hand making american sign language gestures](https://universe.roboflow.com/paul-guerrie-tang1/asl-poly-instance-seg/model/26).

```python
img_url = "https://media.roboflow.com/quickstart/asl_w.jpeg?updatedAt=1678743714769"
universe_model_url = "https://universe.roboflow.com/paul-guerrie-tang1/asl-poly-instance-seg/model/26"

model = roboflow.load_model(model_url=universe_model_url)
pred = model.predict(img_url)
pred.plot()
```
